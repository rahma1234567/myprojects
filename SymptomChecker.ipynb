I decided to create a simple Symptom Checker, where a user inputs a symtpom they are experiencing and the system will suggest possible conditions they might have and advise them on what to do next using a dataset. The three main AI techniques I chose to compare and evaluate are Decision Trees, K-means and Large Language Models. **Decision Trees**

---
Decision Trees are a supervised machine learning algorithm that can be used for regression and classification problems. A symptom checker needs to be a classification model as the symptoms would be classified for different health issues. For a Symptom checker, this AI technique is very suitable because majority of follow up questions are based on yes or no algorithm. For example if the medical chatbot asks 'do you have a headache', the answer will either be 'yes' or 'no', depending on the answer, it will move on to the next question and until its classified as a health condition such as 'flu' or 'migraine'.

*   ***Advantages***

The time required to set up this model is lower compared to other techniques especially because preprocessing only involves labelling the symptoms as a binary or categorical feature which is relatively easy.
This technique is also very fast to implement as it follows an easy algorithm and each result can be traced through a sequence of nodes in the tree, which allows developers to understand how a particular conclusion was reached. Datasets on medical symptoms are likely labeled so it will work well with decision trees.

*   ***Disadvantages***

However, decision trees would be very complicated to use if there are large complicated data, e.g, if it was a Disease Diagnoses System, which could have datasets on images to diagnose certain health conditions, these will make the tree go deeper and deeper causing overfitting.
required will be a questions and symptoms and the outcome data is expected to be the health issue that a patient could potentially have.


**K-Means**

---

K-means is an unsupervised machine learning algorithm which is used to group sets of data into clusters. In a symptom checker chatbot, it can be used to identify patterns in patient symptoms and would be able to group the  patients with similar symptoms.
A Dataset with the symptoms health data will be used and the input of patients symptoms. The symptom checker will group patients similar symptoms, identify the common pattern and assign the users to cluster that would be labeled risk or severity and accordingly decide on the triage of whether its a low, medium or high risk health conditions. the clusters defined for each patients will then be mapped to an action to be taken for example, it will tell patients to book an appointment, seek urgent help or at home care treatments.

*   ***Advantages***

The advantages of using K means is that a labelled data is not required and they can group their data according to their problem requirements.
It can also discover hidden patterns which doctors may have difficulties finding.
New data can be adapted to the clusters without manually changing things around.
It is also easy to understand and implement for the devlopers

*   ***Disadvantages***

The disadvantages <sup>[10]</sup> is that since it works on unlabeled data, the developer must manually create clusters and poor choices might reduce accuracy of the system. Grouping can also be time consuming if the datasets are large. Usually, Medical data is very large and overlapping so when the same prompt is run, it can give different results. Most medical datasets are categroical and unlabeled data is less common so finidng medical data for K-means suitablity will be quite diffcult as K-means usually works the best with numbered data thus this technique will not identify health conditions.


**Large Language Models (LLMs)**

---

LLMs are a deeper subset of AI models that are trained on large amounts of data to understand and reason with human language. They are usually learnt through recognising patterns in languages and they would then predict the most likely word that can be used in a sentence known as a token,  it uses context to generate responses that are relevant and meaningful. As a Symptom Checker , LLMs can be implemented for the purpose of generating responses regarding a patient symptoms. for example if a user asks, 'I have a headache', the LLM will use its training, it will explore and identify symptoms that are common and how severe it is. based on the information it has been taught, the model will generate information response on what they could do for example, drink water, sleep, take paracetamol. It could also generate a question for the patient to answer for example, how long they have had this symptoms, did they take anything to help this problem or whether they have any other symptoms such as nausea. The last output will be the next steps a patient should take for example call 911, or book appointment with a specific  department.

*   ***Advantages***

The advantages of LLM is that one model can handle multiple and complex language tasks such as coding, translating or summarising. They understand context very well and have human like conversations that shows understanding, so people can easily and comfortably get on well with those models. LLM also reduces the developmet time because integration is very easy and wont need to significantly invest on trainig the model or labelling, this makes it ideal for people who want to work fast or have a new business. Another pro is that LLM  can be trained using different languages and regions which makes it a diverse use of technology. There is no end on improvement, each year, new models are being created with better tools and better reasoning and performance.

*   ***Disadvantages***

The disadvantage <sup>[11]</sup> is that training is required,  this can take time depending on the problem requirements, it can create bias in training as well especially as time is evolving and traditional/old advices may not work overtime, thats why it is imporant to regularly retrain the models.

````python
import keras_hub
import keras
import numpy as np
import tensorflow as tf
import pandas as pd
````

First, a GPT-2 model will be created with its preprocessora and backbone using keras_hub: https://keras.io/keras_hub/api/models/gpt2/gpt2_causal_lm/

````python
preprocessor = keras_hub.models.GPT2CausalLMPreprocessor.from_preset(
    "gpt2_base_en",
    sequence_length=128, )

model = keras_hub.models.GPT2CausalLM.from_preset(
    "gpt2_base_en", preprocessor=preprocessor )

backbone_layer = model.get_layer("gpt2_backbone")
````

Before training and fine-tuning is implemented, it's important to check what the ouput of this model is when a symptom is inputted:

Dataset is from Kaggle, Kagglehub has been imported to retrieve the dataset: https://www.kaggle.com/datasets/mdtalhask/ai-powered-symptom-checker-dataset-2025?resource=download

````python
import kagglehub
from kagglehub import KaggleDatasetAdapter

file_path = "AI_Symptom_Checker_Dataset.csv"

df = kagglehub.dataset_load (
  KaggleDatasetAdapter.PANDAS,
  "mdtalhask/ai-powered-symptom-checker-dataset-2025",
  file_path,
)
````

The only columns from the dataset that I will be using is the symptoms, predicted disease and Severity column:

````python
symptoms_clm = "Symptoms"
disease_clm = "Predicted Disease"
severity_clm = "Severity"

df = df[[symptoms_clm, disease_clm, severity_clm ]].copy()

#here you can check the data that I will be using
print(df.dtypes)
df.describe(include="all")
````

After dataset has been retrieved, the data will need to preprocessed and cleaned, these function remove lowercase letters, removes special characters and normalises spacing which is important for trainig and fine-tuning as well as searching for matching symptoms

````python
import re
def clean_text(s: str) -> str:
    s = str(s).lower().strip()
    s = re.sub(r"\s+", " ", s)
    s = re.sub(r"[^a-z0-9,.\- /]", "", s)
    return s

df = df.dropna(subset=[symptoms_clm, disease_clm, severity_clm]).copy()

df[symptoms_clm] = df[symptoms_clm].map(clean_text)
df[disease_clm]  = df[disease_clm].map(clean_text)
df[severity_clm] = df[severity_clm ].map(clean_text)

df = df.drop_duplicates(subset=[symptoms_clm, disease_clm, severity_clm ]).reset_index(drop=True)

print("After cleaning:", df.shape)
print("Num rows:", len(df))
````

To understand more about the dataset, I imported the matplotlib library for the visualisation of the diseases that are common in the dataset

````python
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

common_diseases = df[disease_clm].value_counts().head(15)
plt.figure()
common_diseases.plot(kind="bar")
plt.title("The Common Diseases")
plt.ylabel("Count")
plt.show()
````

Then the rows that are being used will be converted into a trainig text, this converrts structured data into natrual language format, GPT model will learn from full text sentences rather than labels

````python
def row_to_text(row) -> str:
    return (
        f"symptoms: {row[symptoms_clm]}\n"
        f"severity: {row[severity_clm ]}\n"
        f"predicted disease: {row[disease_clm]}"
    )

texts = df.apply(row_to_text, axis=1).tolist()
train_texts, val_texts = train_test_split(texts, test_size=0.1, random_state=42)

BATCH_SIZE = 16
train_ds = (
    tf.data.Dataset.from_tensor_slices(train_texts)
    .shuffle(2000)
    .batch(BATCH_SIZE)
    .prefetch(tf.data.AUTOTUNE)
)

val_ds = (
    tf.data.Dataset.from_tensor_slices(val_texts)
    .batch(BATCH_SIZE)
    .prefetch(tf.data.AUTOTUNE)
)

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=5e-5),
    loss=keras.losses.SparseCategoricalCrossentropy(from_logits=True),
)
#splits data into 90% training and 10% validation - prevents overfitting
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=3 )

print("Train size:", len(train_texts))
print("Val size:", len(val_texts))
print("\n Training:\n", train_texts[0][:300], "...")
````

Now, testing the fine-tuned model

````python
prompt = "i have a headache"
print(model.generate(prompt, max_length=80))
````

Its important to vectorize the symptoms. The TF-IDF technique was used which is a classical NLP technique where symptom text is converted into a numerical vectors, this will act as a fast retrieval layer for LLM.

````python
tfidf = TfidfVectorizer(ngram_range=(1,2), min_df=2)
X = tfidf.fit_transform(df[symptoms_clm].values)
````

This functions will retrieve similiar symptoms which will then return matching symptoms that are relevant as an output. The top-k will select the most unique diseases from the user input, this prevents hallucinations.

````python
def get_symptoms(user_symptom: str, top_k: int = 5):
    user_symptom = clean_text(user_symptom)
    q = tfidf.transform([user_symptom])
    sims = cosine_similarity(q, X).ravel()

    idx_sorted = np.argsort(-sims)

    seen_diseases = set()
    selected = []
    selected_sims = []

    for idx in idx_sorted:
        disease = df.iloc[idx][disease_clm]
        if disease in seen_diseases:
            continue
        seen_diseases.add(disease)
        selected.append(idx)
        selected_sims.append(sims[idx])
        if len(selected) >= top_k:
            break

    out = df.iloc[selected].copy()
    out["similarity"] = selected_sims
    return out
````

now the symptom checker will use all the functions and techniques applied to Retrieves similar cases (TF-IDF), builds a prompt, feeds structured context to GPT-2 and generates a user-friendly response

````python
def symptom_checker(user_symptom: str, top_k: int = 5, max_len: int = 120) -> str:
    cands = get_symptoms(user_symptom, top_k=top_k)

    cand_lines = []
    for i, r in cands.reset_index(drop=True).iterrows():
        cand_lines.append(
            f"{i+1}) symptoms: {r[symptoms_clm]}\n"
            f"   severity: {r[severity_clm]}\n"
            f"   predicted disease: {r[disease_clm]}\n"
        )

    prompt = ( "Please note that I am only an assistant, if you want proper diagnoses, contact GP.\n"
    f"You Said: {clean_text(user_symptom)}\n\n"
        "You may have:\n"
        + "\n".join(cand_lines)
        + "\n\n"
    )

    gen = model.generate(prompt, max_length=max_len)
    return gen
````

See the result of the model using natrual language input:

````python
print(symptom_checker("I have bleeding ", top_k=5))
````

This system combines TF-IDF similarity search for reliable symptom retrieval with a fine-tuned GPT-2 model to generate human readable medical explanation

**Bias and Fairness**<sup>[9][12]</sup>

The AI model is implemented on a Large Language Model (LLM) which is fine tuned on a public dataset from Kaggle. An online public dataset can have existing biases in healthcare data for example under representation of gender, or symptoms biased towards a specific population, these differences can reflect negatively on the results and the system will give uneven guidance among users. To prevent some of the risks, the system was implemented only for a simple checking of symptoms using a dataset rather than diagnosing a medical condition and inventing conditions that may not be true. it provides a simple guidance on what possible conditions you can have.

**Accessibility and Inclusion**

The AI prototype is implemented using python which is known as one of the easiest programming language any software developer will be able to understand what the codes mean. To make it more accessible, notes have been added on many sections of the code explaining what is going on. The output is in simple text format accessible for everyone. Non-technical or non-clincal words have restricted use because not many people will be able to understand. Nevertheless, limitations still remain, the current implementation assumes proficiency in English and access to internet and digital devices which may not be suitable for non-English speakers, people with visual impairment or those who have limited access to digital devices. It is also a text based interaction which may not be accessible for user with cognitive or reading difficulties.


**Ethical Implications**<sup>[9][12]</sup>

There is a high level of ethical risk for deploying AI in a medical environment particularly when a system gives medical advices or diagnoses a condition. Incorrect or misleading could cause unnecessary concerns and anxiety to the user or they would believe and delay in seeking further guidance from a professional healthcare. My system checker prototype reduces some ethical risks by avoiding definite diagnosing and explicitly adding a disclaimer that further encourages a proper diagnosis from a healthcare professional and not to rely wholly on this system.
This prototype is not intended for a real-world clinician deployment and the output should only be treated as a possibility condition based on the dataset being used rather than an authoritative medical diagnosis.

**Sustainability Considerations:**

The model used in the prototype is GPT2 which is usually smaller than most LLM models such as Gemma but even GPT2 still requires large computational resource for fine-tuning and inference. Some suggestions for reducing environmental impact in production deployment is the use of smaller LLM models because it would use less processing power and memory.
