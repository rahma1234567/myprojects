This Jupyter Notebook file was created to understand and visualise the information in the dataset including columns, rows and data types. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("bank_transactions_data_2.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
print (data.shape)
```

```python
# Check for missing values
print("Missing values per column:")
print(data.isnull().sum())
print(f"\nTotal missing: {data.isnull().sum().sum()}")
```

```python
# Check for duplicates
print(f"\nDuplicate rows: {data.duplicated().sum()}")
```

```python
# Scatter Graph of transaction amount and the balance in the account
plt.figure()
plt.scatter(data["AccountBalance"], data["TransactionAmount"], alpha=0.5)
plt.xlabel("Account Balance")
plt.ylabel("Transaction Amount")
plt.title("Transaction Amount vs Account Balance")
plt.show()
```

```python
attempts = data[data['LoginAttempts'] >= 3]
print(attempts)
```

```python
plt.figure(figsize=(10,5))
plt.bar(attempts["AccountID"].astype(str), attempts["LoginAttempts"])

plt.xlabel("AccountID")
plt.ylabel("LoginAttempts")
plt.title("Accounts with 3 or more login attempts")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
